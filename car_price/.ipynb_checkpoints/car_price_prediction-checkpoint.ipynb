```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('car data.csv')
```

```python
df.head()
```

```python
final_df = df[['Year', 'Selling_Price', 'Present_Price', 'Kms_Driven',
       'Fuel_Type', 'Seller_Type', 'Transmission', 'Owner']]
```

```python
final_df.head()
```

```python
final_df['Current_Year'] = 2023
```

```python
final_df.head()
```

```python
final_df['Years_Used'] = final_df['Current_Year'] - final_df['Year']
```

```python
final_df.head()
```

```python
final_df.drop(['Year', 'Current_Year'], axis=1, inplace=True)
```

```python
final_df.head()
```

```python
final_df =pd.get_dummies(final_df, drop_first=True)
```

```python
final_df
```

```python
columns_to_convert = final_df.columns[final_df.dtypes == bool]
final_df[columns_to_convert] = final_df[columns_to_convert].astype('int')
```

```python
final_df.head()
```

```python
final_df.corr()
```

```python
corr_mat = final_df.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_mat, annot=True, cmap='coolwarm', center=0)
plt.title("Correlation Heatmap")
plt.show()
```

```python
X = final_df.iloc[:,1:]
Y = final_df.iloc[:,0]
```

```python
Y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3, random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
 #Randomized Search CV

n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]
max_features = ['sqrt', 'log2']
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]
min_samples_split = [2, 5, 10, 15, 100]
min_samples_leaf = [1, 2, 5, 10]
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}

print(random_grid)
```

```python
rf = RandomForestRegressor()
```

```python
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid,scoring='neg_mean_squared_error', n_iter = 10, cv = 5, verbose=2, random_state=42, n_jobs = 1)
```

```python
rf_random.fit(X_train,Y_train)
```

```python
rf_random.best_params_
```

```python
rf_random.best_score_
```

```python
predictions = rf_random.predict(X_test)
predictions[:5]
```
